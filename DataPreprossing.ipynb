```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
preData = pd.read_csv('Stock_Currency_Commodity.csv')
print(preData.shape)
preData = preData.dropna(thresh=7)   #drop if at least 7 values are NOT Null.. Half of Features
preData = preData.dropna(subset=['NDAQ'])  #Dropping column if NDAQ has NaN value
print(preData.shape)
preData = preData.round({'NDAQ':2})
#preData.to_csv('cleanData.csv')
#Then added class labe from excel by checking NDAQ's next day value: 1 if increase, -1 if decrease, 0 if equal
headerLen = len(list(preData))  #15 Feture
```

```python
#Generating Class Labels for increase gives 1 and decrease gives -1.
ndaq = preData['NDAQ']
ndaq0 = ndaq[:ndaq.shape[0]-1]
ndaq1 = ndaq[1:]
print(ndaq0.shape)
print(ndaq1.shape)
diff = np.subtract(ndaq1,ndaq0)
indexPos = np.where(diff>0)
ones  = np.ones(len(indexPos))
np.put(diff,indexPos,ones) 
indexNeg =  np.where(diff<0)
minusOne = np.negative(np.ones(len(indexNeg)))
np.put(diff,indexNeg,minusOne)
```

```python
preData.drop(preData.tail(1).index,inplace=True)  #Dropping last row as it have no increasing decreasing value
preData = preData.set_index(pd.Index(list(range(0,preData.shape[0])))) #Setting index as remove rows remove indexes also
clss = pd.DataFrame(diff)
clss = clss.rename(columns={'NDAQ':'Target'})
clss = clss.set_index(pd.Index(list(range(0,clss.shape[0]))))
dataLabel = pd.concat([preData,clss],axis=1)
print(dataLabel)
#dataLabel.to_csv('cleanDataLabel.csv')
```
